```python
fig, ax = plt.subplots(round(len(df_numeric) / 3), 3, figsize = (18, 12))

for i, ax in enumerate(fig.axes):
    if i < len(df_numeric) - 1:
        sns.regplot(x=df_numeric[i],y='Credit_limit', data=df_numeric, ax=ax)
```

```python
corr = df_numeric.drop('CLIENTNUM', axis=1).corr() # We already examined SalePrice correlations
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```
